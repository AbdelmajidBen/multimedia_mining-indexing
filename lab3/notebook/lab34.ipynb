```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/Users/mac/Desktop/MST AISD/S3/mutimedia mining/multimedia-mining-indexing labs/lab3/images/im7.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

r, g, b = cv2.split(image)

hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
plt.figure(figsize=(10, 5))
plt.title('Histogramme des couleurs')
plt.xlabel('Intensit des pixels')
plt.ylabel('Nombre de pixels')
plt.xlim([0, 256])
plt.plot(hist_r, color='red', label='Rouge')
plt.plot(hist_g, color='green', label='Vert')
plt.plot(hist_b, color='blue', label='Bleu')
plt.legend()
plt.show()
```
